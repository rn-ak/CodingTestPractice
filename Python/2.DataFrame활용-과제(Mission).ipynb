## Pandas참조 후 titanic 데이터를 가져오기
- name열이 index로 설정

```python
import pandas as pd
```

```python
titanic = pd.read_excel('titanic3.xls', index_col='name')
titanic.head()
```

## 데이터프레임의 columns만 조회

```python
titanic.columns
```

## .drop()을 이용해서 2개의 열을 삭제
- home.dest
- embarked
- sibsp
- 원본에 반영

```python
# axis=1
titanic_drop = titanic.drop(['home.dest', 'embarked', 'sibsp'], axis=1)
titanic_drop
```

## del을 이용한 열 삭제
- parch

```python
del titanic['parch']
```

## 사망자 명단 조회하기

```python
dead_df = titanic['survived'] != 1 
```

```python
titanic[dead_df]
```

## 시신 식별번호(body)가 없는 데이터 조회하기

```python
# isnull
body_df = titanic['body'].isnull()
```

```python
titanic[body_df]
```

살아있는 사람은 시신 식별번호가 없는 것은 당연

## 사망자이나 시신식별번호가 없는 명단 찾기

```python
titanic[dead_df | body_df]
```

사망자가 809명인데, 식별번호가 없는 명단이 688명이나 된다.

## 사망자이나 시신식별번호가 없는 명단으로 새로운 df 만들기
- `notfound_list`

```python
notfound_list = titanic[dead_df & body_df]
notfound_list
```

## notfound_list의 승객 등급별 개수 확인

```python
notfound_list['pclass'].value_counts()
```

3등급(하류층)이 많으나, 이건 당연할 수 있다. 비율을 구해보자.

## 전체 titanic 전체 승객 등급별 개수 확인

```python
titanic['pclass'].value_counts()
```

## 승객 등급별 실종 비율 확인

```python
notfound_list['pclass'].value_counts() / titanic['pclass'].value_counts()
```

## 1. 사망자의 하류층 티겟 가격 분포와 2. 생존자의 하류층 티켓 가격 분포를 확인해보자.

```python
pclass_3 = titanic['pclass'] == 3
alive = titanic['survived'] == 1
dead = titanic['survived'] == 0
```

```python
titanic[pclass_3 & dead].describe()
```

```python
titanic[pclass_3 & alive].describe()
```

3등급에서 티켓가격의 차이가 생존유무와 관련성이 있어보이지 않는다.
