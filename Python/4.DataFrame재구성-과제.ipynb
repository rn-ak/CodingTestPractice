```python
import pandas as pd
```

## Pandas참조 + students 데이터프레임 생성

```python
students = pd.read_csv('sample1.csv')
students
```

## '총점' 열을 데이터프레임에 생성하고 값을 계산

```python
students['총점'] = students['국어점수'] + students['수학점수'] + students['영어점수']
students
```

## 변수에 그룹화 데이터된 데이터 저장
- '반'과 '석식 신청'을 기준으로

```python
students_groupby = students.groupby(['반','석식 신청'])
students_groupby
```

```python
students_groupby.groups
```

## 반-석식신청으로 그룹화된 학생 그룹의 '총점' 분석
- agg()를 이용해보자
- count, median, max, min, mean

```python
analysis = students_groupby.agg({'총점': ['count', 'median', 'max', 'min', 'mean']})
analysis
```

## 총점의 '최저점 - 평균'이 낮은 순으로 정렬하기

- 석식 신청을 안한 친구들의 성적 최저점이 낮은 편인가?

```python
analysis['총점'].sort_values(by=['min', 'mean'])
```

## 전체 인원에서 석식 여부가 성적에 영향이 있는지 분석해보기
- 석식 신청으로 그룹화하고
- count median max min mean std를 확인해보자

```python
s_g = students.groupby('석식 신청')
```

```python
s_g.agg({'총점': ['count', 'median', 'max', 'min', 'mean', 'std']})
```

## 부모 학업 성취 수준에 따라 성적에 영향이 있는지 분석해보기
- 학업 성취 수준으로 그룹화하고
- mean count std median max min 을 확인해보자
- 평균 높은 순으로 정렬

```python
s_g2 = students.groupby('부모 학업 성취')
```

```python
parents = s_g2.agg({'총점' : ['mean', 'count', 'std', 'median', 'max', 'min']})
parents
```

```python
parents.sort_values(('총점', 'mean'), ascending=False)
```
