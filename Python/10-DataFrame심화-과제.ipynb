## 수익을 다음으로 분류
- 60000이상 고임금
- 30000이상~60000미만 평균임금
- 30000미만 저임금
- 열 이름은 'Class_Income'

```python
def classify_income(Income) :
    if Income >= 60000 :
        return '고임금'
    elif 60000 > Income >= 30000:
        return '평균임금'
    elif 30000 > Income :
        return '저임금'
    return None
```

```python
markting_df['Class_Income'] = markting_df['Income'].apply(classify_income)
```

## Alone, Absurd, YOLO를 Single로 변경
1. reset_index()로 되돌리고
2. replace으로 값을 바꾼다음에
3. 다시 set_index()로 설정해보자.
4. 실제로 잘 변경되었는지 get_level_values(0)후 value_counts()로 데이터 경우의 수를 확인해보자.

```python
markting_df.reset_index(inplace=True)
markting_df
```

```python
markting_df['Marital_Status'].replace({'Alone':'Single', 'Absurd':'Single', 'YOLO':'Single'}, inplace=True)
markting_df
```

```python
markting_df.set_index(['Marital_Status', 'Education'], inplace=True)
markting_df.sort_index(inplace=True)
markting_df
```
